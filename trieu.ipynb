```python
import cv2
import numpy as np
import os

def createVariants(imagePath, outputFolder):
    if not os.path.exists(outputFolder):
        os.makedirs(outputFolder)
    
    baseName = os.path.basename(imagePath).split('.')[0]
    img = cv2.imread(imagePath)
    
    if img is None:
        return
        
    cv2.imwrite(os.path.join(outputFolder, f"{baseName}_original.jpg"), img)
    
    height, width = img.shape[:2]
    center = (width / 2, height / 2)
    matrix = cv2.getRotationMatrix2D(center, 7, 1)
    rotatedImg = cv2.warpAffine(img, matrix, (width, height))
    cv2.imwrite(os.path.join(outputFolder, f"{baseName}_rotated.jpg"), rotatedImg)
    
    noise = np.random.normal(0, 15, img.shape).astype(np.uint8)
    noisyImg = cv2.add(img, noise)
    cv2.imwrite(os.path.join(outputFolder, f"{baseName}_noisy.jpg"), noisyImg)
    
    brightImg = cv2.convertScaleAbs(img, alpha=1.1, beta=20)
    cv2.imwrite(os.path.join(outputFolder, f"{baseName}_bright.jpg"), brightImg)

imageFiles = [
    r"D:\HocTangCuong\conmeo.jpg", 
    r"D:\HocTangCuong\hoa.jpg", 
    r"D:\HocTangCuong\images.jpg", 
    r"D:\HocTangCuong\ngoinha.jpg"
]

uniqueFiles = list(set(imageFiles))

for path in uniqueFiles:
    createVariants(path, "lab4_dataset")
```
